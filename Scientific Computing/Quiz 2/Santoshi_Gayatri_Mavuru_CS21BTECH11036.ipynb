```python
# Question 1

import numpy as np
import pandas as pd
T = -1*np.eye(10,10,k=-1)+np.eye(10,10)*2+-1*np.eye(10,10,k=1)

n = len(T)
L = np.eye(10)
U = np.zeros((n,n))

U[0][0] = T[0][0]

for i in range(1,n):
    U[i-1][i] = T[i-1][i]
    L[i][i-1] = T[i][i-1]/U[i-1][i-1]
    U[i][i] = T[i][i]-L[i][i-1]*T[i-1][i]

T_df = pd.DataFrame(T)
U_df = pd.DataFrame(U)
L_df = pd.DataFrame(L)

print("Tridiagonal Matrix T : \n")
print(T_df.to_string(index=False, header=False))
print("\nL : \n")
print(L_df.to_string(index=False, header=False))
print("\nU : \n")
print(U_df.to_string(index=False, header=False))
```

```python
# Question 4

def mat_inv(m):

    n = len(m)
    aug_mat = []
    for i in range(n):
        row = m[i]
        aug_row = []
        
        for j in range(n):
            aug_row.append(row[j])
        
        for j in range(n):
            aug_row.append(1 if i == j else 0)
        aug_mat.append(aug_row)

    for i in range(n):
        diag_el = aug_mat[i][i]
        
        for j in range(2 * n):
            aug_mat[i][j] /= diag_el

        for k in range(n):
            if k != i:
                factor = aug_mat[k][i]
                for j in range(2 * n):
                    aug_mat[k][j] -= factor*aug_mat[i][j]

    inv = []
    for row in aug_mat:
        inv.append(row[n:])

    return inv

m = [
    [2,0,3,0],
    [0,2,0,0],
    [1,0,2,0],
    [0,1,3,4]
]

x = mat_inv(m)
for row in x:
    print(row)
```

```python
# Question 6

def Queens(n):

    def can_put(r, c):
        for i in range(r):
            if board[i] == c or board[i] - i == c - r or board[i] + i == c + r: # conditions for clash
                return False
        return True

    def put(n, r):
        if r == n:
            solns.append(board[:])
            return
        for c in range(n):
            if can_put(r, c):
                board[r] = c
                put(n, r + 1)
                board[r] = 0

    solns = []
    board = [0] * n
    put(n, 0)
    ans = []
    for sol in solns:
        x = []
        for i in sol:
            x.append("." * i + "Q" + "." * (n - i - 1))
        ans.append(x)

    return ans
n = int(input("Enter n : "))
print("Possible Solutions are : ")
print(Queens(n))
```

```python
# Question 7

def paren(ans,n,s,open,close):
    if close == n:
        ans.append(s)
        return
    
    else:
        if open < n:
            s += '('
            paren(ans,n,s,open+1,close)
            s = s[:-1]
        if open > close:
            s += ')'
            paren(ans,n,s,open,close+1)


def all_paren(n):
    ans = []
    s = ''
    paren(ans,n,s,0,0)
    return ans            

n = int(input("Enter n : "))
print("Possible pairs are : ")
print(all_paren(n))
```
